# U.S. Oil and Gas Production Analysis

Analysing the [Kaggle Dataset](https://www.kaggle.com/djzurawski/us-oil-and-gas-production-june-2008-to-june-2018) with information about Oil and Gas production in the US from June 2008 to June 2018.

## Imports

```python
import numpy as np
import pandas as pd
import matplotlib
```

## Let's load the Gas production:

```python
file = "data/U.S._natural_gas_production.csv"
gas_df = pd.read_csv(file, decimal=",")
gas_df.head(3)
```

```python
# Check how many rows and columns in the dataframe
gas_df.shape
```

```python
# Display all the available column names in the dataframe
gas_df.columns
```

```python
# Display additional info about each columns such as data types and number of non-null values
gas_df.info()
```

## Convert Month to Datetime object

```python
gas_df['Month'] = pd.to_datetime(gas_df['Month'])
```

### Check your code

```python
month_type = gas_df['Month'].dtype

from nbresult import ChallengeResult

result = ChallengeResult('date',
    month_type=month_type,
)
result.write()

print(result.check())
```

```python
gas_df['Month'].dt.year.head()
```

```python
gas_df['Month'].dt.month.tail()
```

## Yearly Gas production

```python
yearly_gas_df = gas_df.groupby(gas_df['Month'].dt.year).sum()
```

```python
yearly_gas_df.head()
```

### Check your code

```python
index_year = yearly_gas_df.index[0]
yearly_gas_shape = yearly_gas_df.shape
us_total = yearly_gas_df.iloc[0,0]

from nbresult import ChallengeResult

result = ChallengeResult('full_gas',
    index_year=index_year,
    yearly_gas_shape=yearly_gas_shape,
    us_total=us_total
)
result.write()

print(result.check())
```

## Plotting Results

```python
yearly_gas_df.filter(items=['U.S.']).plot(xlabel='Year', title='U.S. Gas Production by year');
```

```python
yearly_gas_df.filter(items=['U.S.']).plot(kind='bar', xlabel='Year', title='U.S. Gas Production by year');
```

## Filtering by boolean indexing

```python
yearly_gas_df.index
```

```python
np.logical_and(yearly_gas_df.index > 2008, yearly_gas_df.index < 2018)
```

```python
filtered_yearly_gas_df = yearly_gas_df[np.logical_and(yearly_gas_df.index > 2008, yearly_gas_df.index < 2018)]
filtered_yearly_gas_df
```

```python
filtered_yearly_gas_df.filter(items=['U.S.']).plot(xlabel='Year', title='U.S. Gas Production by year');
```

```python
filtered_yearly_gas_df.filter(items=['U.S.']).plot(kind='bar', xlabel='Year', title='U.S. Gas Production by year');
```

### Check your code

## Comparing with Crude Oil Production

```python
oil_df = pd.read_csv('data/U.S._crude_oil_production.csv', parse_dates=['Month'])
oil_df.head()
```

```python
oil_df.info()
```

## Yearly Oil production

```python
yearly_oil_df = oil_df.groupby(oil_df['Month'].dt.year).sum()
yearly_oil_df
```

```python
yearly_oil_df.columns = yearly_oil_df.columns.str.strip()
```

```python
yearly_oil_df.columns
```

```python
yearly_oil_df.filter(items=['U.S. Crude Oil']).plot(kind='bar');
```

```python
filtered_yearly_oil_df = yearly_oil_df[np.logical_and(yearly_oil_df.index > 2008, yearly_oil_df.index < 2018)]
```

### Check your code

```python
from nbresult import ChallengeResult

result = ChallengeResult('oil',
    filtered_oil_shape=filtered_yearly_oil_df.shape,
    filtered_oil_index_year=filtered_yearly_oil_df.index[0],
    us_total=filtered_yearly_oil_df.iloc[0,0]
)
result.write()

print(result.check())
```

## Merging Oil and Gas Production

```python
total_gas = filtered_yearly_gas_df['U.S.']
total_gas.rename(columns={'U.S.': 'Gas'})
# total_oil = filtered_yearly_oil_df['U.S. Crude Oil'].rename(columns={'U.S. Crude Oil': 'Crude Oil'})
```

```python
total_gas = pd.DataFrame(filtered_yearly_gas_df['U.S.']).rename(columns={'U.S.': 'Gas'})
total_oil = pd.DataFrame(filtered_yearly_oil_df['U.S. Crude Oil']).rename(columns={'U.S. Crude Oil': 'Crude Oil'})
```

```python
total_gas
```

```python
merged_df = pd.concat([total_oil, total_gas], axis='columns')
merged_df
```

```python
plot = merged_df.plot(kind="bar")
plot.set_xlabel("Year")
plot.legend(['Gas (Millions of Cubic feet)', 'Crude Oil (Thousands of barrels)']);
```
